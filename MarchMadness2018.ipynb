```python
def combined_data(team_id, year, tw = 1): 
    r = reg_season_data(team_id, year)
    t = tor_season_data(team_id, year)
    
    if 
    
    if t == [0]:
        t = [0] * len(r)
    
    
    total_num = r[0] + t[0]
    
    len_gamesWon = r[1] + tw*t[1]
    len_gamesLose = r[2] + tw*t[2]
    
    score_per_game = (r[3] + tw*t[3]) / total_num
    score_allowed_per_game = (r[4] + tw*t[4]) / total_num
    num_3pt_made_per_game = (r[5] + tw*t[5]) / total_num
    num_3pt_attempt_per_game = (r[6] + tw*t[6]) / total_num
    opnt_3pt_made_per_game = (r[7] + tw*t[7]) / total_num
    opnt_3pt_attempt_per_game = (r[8] + tw*t[8]) / total_num
    num_FG_made_per_game = (r[9] + tw*t[9]) / total_num
    num_FG_attempt_per_game = (r[10] + tw*t[10]) / total_num
    opnt_FG_made_per_game = (r[11] + tw*t[11]) / total_num
    opnt_FG_attempt_per_game = (r[12] + tw*t[12]) / total_num
    turnover_per_game = (r[13] + tw*t[13]) / total_num
    opnt_turnover_per_game = (r[14] + tw*t[14]) / total_num
    assist_per_game = (r[15] + tw*t[15]) / total_num
    ofns_rebound_per_game = (r[16] + tw*t[16]) / total_num
    opnt_ofns_rebound_per_game = (r[17] + tw*t[17]) / total_num
    defn_rebound_per_game = (r[18] + tw*t[18]) / total_num
    opnt_defn_rebound_per_game = (r[19] + tw*t[19]) / total_num
    foul_per_game = (r[20] + tw*t[20]) / total_num
    opnt_foul_per_game = (r[21] + tw*t[21]) / total_num
    free_throw_made_per_game = (r[22] + tw*t[22]) / total_num
    free_throw_attempt_per_game = (r[23] + tw*t[23]) / total_num
    opnt_free_throw_made_per_game = (r[24] + tw*t[24]) / total_num
    opnt_free_throw_attempt_per_game = (r[25] + tw*t[25]) / total_num
    
    
    
#   % of turn over 
    pct_won_turn_over_rate = (100*turnover_per_game)/(num_FG_attempt_per_game+\
                                                   (0.44*free_throw_attempt_per_game)+ turnover_per_game)

  
    
#   % of effective field goal 
    pct_effective_FG = (num_FG_made_per_game + 0.5*num_3pt_made_per_game)/num_FG_attempt_per_game 
    
    
#   posessions 
    possession = 0.5*((num_FG_attempt_per_game+0.4*free_throw_attempt_per_game-1.07*\
                       (ofns_rebound_per_game/(ofns_rebound_per_game+opnt_defn_rebound_per_game))*\
                       (num_FG_attempt_per_game - num_FG_made_per_game)+turnover_per_game)+\
                      (opnt_FG_attempt_per_game+0.4*opnt_free_throw_attempt_per_game-1.07*\
                       (opnt_ofns_rebound_per_game/(opnt_ofns_rebound_per_game + defn_rebound_per_game))*\
                       (opnt_FG_attempt_per_game - num_FG_made_per_game)+opnt_turnover_per_game))
    
    return [len_gamesWon, len_gamesLose, score_per_game, score_allowed_per_game, num_3pt_made_per_game,\
            num_3pt_attempt_per_game, opnt_3pt_made_per_game, opnt_3pt_attempt_per_game,\
            num_FG_made_per_game, num_FG_attempt_per_game, opnt_FG_made_per_game, opnt_FG_attempt_per_game,\
            turnover_per_game, opnt_turnover_per_game, assist_per_game, \
            ofns_rebound_per_game, opnt_ofns_rebound_per_game ,defn_rebound_per_game, opnt_defn_rebound_per_game,\
            foul_per_game, opnt_foul_per_game, free_throw_made_per_game, free_throw_attempt_per_game,\
            opnt_free_throw_made_per_game, opnt_free_throw_attempt_per_game, pct_won_turn_over_rate,\
            pct_effective_FG, possession]


combined_data(1276,2017)
    
    
```

```python
def diffvec(team1, team2, year):
    w = np.array(combined_data(team1, year))
    l = np.array(combined_data(team2, year))
    t = w-l
    
    return t

diffvec(1276, 1277, 2017)
```
