```python
def reg_season_data(team_id, year):
    
    
#bring year data
    year_data_compact_pd = reg_season_compact_pd[reg_season_compact_pd['Season'] == year]
    year_data_detailed_pd = reg_season_detailed_pd[reg_season_detailed_pd['Season'] == year]
    # detailed num game, compact num game 같은것 확인함
    ##numgames = len(year_data_compact_pd)
    #by number of won/lose
    gamesWon_compact = year_data_compact_pd[year_data_compact_pd.WTeamID == team_id] 
    gamesLose_compact = year_data_compact_pd[year_data_compact_pd.LTeamID == team_id]
    gamesWon = year_data_detailed_pd[year_data_detailed_pd.WTeamID == team_id] 
    gamesLose = year_data_detailed_pd[year_data_detailed_pd.LTeamID == team_id]
    #print(gamesWon)
    len_gamesWon_compact = len(gamesWon_compact)
    len_gamesLose_compact = len(gamesLose_compact)
    totalgame_compact = len_gamesWon_compact + len_gamesLose_compact
#     print(totalgame_compact)
    
    len_gamesWon = len(gamesWon)
    len_gamesLose = len(gamesLose)
    totalgame = len_gamesWon + len_gamesLose
    
#     print(reg_totalgame)
    #winning ratio
    winning_ratio = len_gamesWon/totalgame
    numgames = totalgame
#     print(winning_ratio)
    


    #by point(avg score per game/ total score/ total score won / total score lose)
    won_score_total = gamesWon_compact['WScore'].sum()
    lose_score_total = gamesLose_compact['LScore'].sum()
    
    opnt_won_score_total = gamesWon_compact['LScore'].sum()
    opnt_lose_score_total = gamesLose_compact['WScore'].sum()
    total_score = won_score_total + lose_score_total
    avg_score_per_game = total_score/numgames
    avg_score_allowed_to_opont = (opnt_won_score_total+ opnt_lose_score_total)/numgames 
#     print(avg_score_allowed_to_opont)

        #3point made
            
    num_of_won_3point_made = gamesWon['WFGM3'].sum()
    num_of_won_3point_attempt = gamesWon['WFGA3'].sum()
    num_of_lose_3point_made = gamesLose['LFGM3'].sum()
    num_of_lose_3point_attempt = gamesLose['LFGA3'].sum()
#     print(num_of_won_3point_made)
    num_of_3point_made_total = num_of_won_3point_made + num_of_lose_3point_made
    num_of_3point_attempt_total = num_of_won_3point_attempt + num_of_lose_3point_attempt
#     print(num_of_3point_made_total)
    # num of 3 per game 
    num_of_3point_made_per_game = num_of_3point_made_total/numgames
    num_of_3point_attempt_per_game = num_of_3point_attempt_total/numgames 
#     print(num_of_3point_made_per_game)
    # % of 3 made per game
    prcnt_of_3_point_made = num_of_3point_made_per_game/num_of_3point_attempt_per_game
#     print(prcnt_of_3_point_made)      
            
    
    #field goal made
    num_of_won_FG_made = gamesWon['WFGM'].sum()
    num_of_won_FG_attempt = gamesWon['WFGA'].sum()
    num_of_lose_FG_made = gamesLose['LFGM'].sum()
    num_of_lose_FG_attempt = gamesLose['LFGA'].sum()
    
    num_of_FG_made_total = num_of_won_FG_made + num_of_lose_FG_made
    num_of_FG_attempt_total = num_of_won_FG_attempt + num_of_lose_FG_attempt
        #opponent
    num_opnt_FG_attempt_total = gamesLose['WFGA'].sum()+gamesWon['LFGA'].sum()
    num_opnt_FG_made_total = gamesLose['WFGM'].sum()+gamesWon['LFGM'].sum()
    
    #num of FG per game 
    num_of_FG_made_per_game = num_of_FG_made_total/numgames
    num_of_FG_attempt_per_game = num_of_FG_attempt_total/numgames 

    #% of FG made 
    prcnt_of_FG_point_made = num_of_FG_made_per_game/num_of_FG_attempt_per_game
    
    #by turnover(turnover per game)
    
        #id team 
    won_turnover= gamesWon['WTO'].sum()
    lose_turnover = gamesLose['LTO'].sum()
    turnover_total = won_turnover + lose_turnover

    avg_turnover = turnover_total/numgames
#     print(avg_turnover)
        #opponent team  ####체크
    opnt_lose_turnover = gamesLose['WTO'].sum()
    opnt_won_turnover = gamesWon['LTO'].sum()

    opnt_turnover_total = opnt_lose_turnover + opnt_won_turnover
    opnt_turnover_avg = opnt_turnover_total/numgames
#     print(opnt_turnover_avg)

    
    #assist per game 
    assist_won = gamesWon['WAst'].sum()
    assist_lose = gamesLose['LAst'].sum()
    
    assist_total = assist_won + assist_lose
    assist_avg = assist_total/numgames
#     print(assist_avg)

#rebound per game
    
        #offensive
    ofns_rebound_won =gamesWon['WOR'].sum()
    ofns_rebound_lose = gamesLose['LOR'].sum()
    
    ofns_rebound_total = ofns_rebound_won+ ofns_rebound_lose
    ofns_rebound_avg = ofns_rebound_total/numgames
            #opponemt
    opnt_ofns_rebound_lose = gamesLose['WOR'].sum()
    opnt_ofns_rebound_won = gamesWon['LOR'].sum()
    opnt_ofns_rebound_total= opnt_ofns_rebound_won + opnt_ofns_rebound_lose
    
        #defensive
    defn_rebound_won =gamesWon['WDR'].sum()
    defn_rebound_lose = gamesLose['LDR'].sum()
    
    defn_rebound_total= defn_rebound_won+defn_rebound_lose
            #opponent
    opnt_defn_rebound_lose = gamesLose['WDR'].sum()
    opnt_defn_rebound_won = gamesWon['LDR'].sum()
   
    opnt_defn_rebound_total=opnt_defn_rebound_lose+opnt_defn_rebound_won
    
    opnt_defn_rebound_total = opnt_ofns_rebound_won + opnt_defn_rebound_lose
    opnt_defn_rebound_avg = opnt_ofns_rebound_total/numgames
#     print(opnt_defn_rebound_avg)
    
    #opnt foul 
    opnt_lose_foul = gamesLose['WPF'].sum()
    opnt_won_foul = gamesWon["LPF"].sum()
        
    opnt_foul_total = opnt_lose_foul + opnt_won_foul
    opnt_foul_avg = opnt_turnover_total/numgames
    
    # free throw
    won_free_throw_attempt = gamesWon['WFTA'].sum()
    won_free_throw_made = gamesWon['WFTM'].sum()
    lose_free_throw_attempt = gamesLose['WFTA'].sum()
    lose_free_throw_made = gamesLose['WFTM'].sum()
    
    free_throw_attempt_total = won_free_throw_attempt + lose_free_throw_attempt
    free_throw_made_total = won_free_throw_made + lose_free_throw_made
   
    opnt_free_throw_attempt_total = gamesLose['WFTA'].sum()+gamesWon['LFTA'].sum()
    opnt_free_throw_made_total= gamesLose['WFTM'].sum()+gamesWon['LFTM'].sum()
    
    #% of free throw made
    
    per_free_throw_made = free_throw_made_total/free_throw_attempt_total
#     print(per_free_throw_made)
    
    
    #% of turn over 
    per_won_turn_over_rate = (100*turnover_total)/(num_of_FG_attempt_total+(0.44*free_throw_attempt_total)+ turnover_total)

#     print(per_won_turn_over_rate)
    
    #% of effective field goal 
    per_effective_FG = (num_of_FG_made_total+0.5*num_of_3point_made_total)/num_of_FG_attempt_total
#     print(per_effective_FG)
    
    #posessions 
    possession = 0.5*((num_of_FG_attempt_total+0.4*free_throw_attempt_total\
                       -1.07*(ofns_rebound_total/(ofns_rebound_total+opnt_defn_rebound_total))\
                       *(num_of_FG_attempt_total-num_of_FG_made_total)+turnover_total)\
                      +(num_opnt_FG_attempt_total+0.4*opnt_free_throw_attempt_total\
                        -1.07*(opnt_ofns_rebound_total/(opnt_ofns_rebound_total+defn_rebound_total))\
                        *(num_opnt_FG_attempt_total-num_opnt_FG_made_total)+opnt_turnover_total))
#     print(possession)
    
    return [gamesWon, gamesLose, winning_ratio, won_score_total,\
            lose_score_total, opnt_won_score_total, opnt_lose_score_total,avg_score_per_game,\
           num_of_3point_made_per_game,num_of_3point_attempt_per_game,prcnt_of_3_point_made,\
           num_of_FG_made_per_game, num_of_FG_attempt_per_game,prcnt_of_FG_point_made,turnover_total,\
            avg_turnover,opnt_turnover_total,opnt_turnover_avg,assist_total,assist_avg,defn_rebound_total,\
            opnt_defn_rebound_total,opnt_defn_rebound_avg,opnt_foul_avg,opnt_foul_total,\
            opnt_free_throw_attempt_total,opnt_free_throw_made_total,per_free_throw_made,\
            per_won_turn_over_rate,per_effective_FG,possession]
reg_season_data(1276,2017)
# def tor_data_game_results(team_id, year): 
#     #tournament seed(how are we going to weight is important)
#     tor_seeds = tor_seeds_pd['Seed']






    

    

    

```
